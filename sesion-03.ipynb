## Reto 01: Sistema de reservaciones

### 🎯 Objetivo
⚒️ Desarrollar un sistema simplificado de gestión de reservaciones para un restaurante utilizando listas de tuplas.

### 📝 Instrucciones
📜 Define una lista de tuplas donde cada tupla represente una reservación. Con los siguientes datos:

Nombre del cliente
Cantidad de personas
➕ Añade al menos 5 reservaciones a la lista.

➖ Elimina una reservación existente de la lista.

🔄 Finalmente recorre la lista de reservaciones e imprime un mensaje personalizado:

✉️ Reservación: Mario Gonzalez, Para: 4 personas.
### ✅ Desafío adicional: 
Piensa en cómo podrías mejorar la gestión de las reservaciones. Algunas ideas pueden incluir manejar situaciones donde intentas eliminar una reservación que no existe o cómo manejar la actualización de una reservación existente.



```python
reservaciones = [
    ("Mario Gonzalez", 4),
    ("Maria Cuadros", 2),
    ("Sarah Barrientos", 2),
    ("Samuel López", 5),
    ("Elena Garro", 10)
]

reservaciones.append(("PatoTective", 7))
reservaciones.append(("Frida Martinez", 6))
reservaciones.remove(("Mario Gonzalez", 4))

for nombre, personas in reservaciones: 
    print(f"Reservación: {nombre}, Para: {personas} personas.")

```
