## Reto 02: Gestión de préstamos en biblioteca

### 🎯 Objetivo
⚒️ Desarrollar un sistema simplificado de gestión de préstamos de libros en una biblioteca, conjuntos y diccionarios.

### 📝 Instrucciones
📚 Define un diccionario con al menos 5 libros disponibles. Con los siguientes datos:
Puedes usar una clave alfanumérica para cada libro.
Cada clave debe contener un diccionario que contenga:
título
autor
año
📖 Crear un conjunto vacío que almacenara los libros que serán prestados.

➕ Prestar un libro:

Verificar si el libro ya está prestado.
Si no está prestado, añadirlo al conjunto de libros prestados, y mostrar un mensaje personalizado.
✉️ Libro 'The Great Gatsby' prestado con éxito.
# o en caso contrario...
✉️ Este libro ya está prestado.
🔄 Devolver un libro:

Verificar si el libro está prestado.
Si está prestado, eliminarlo del conjunto de libros prestados, y mostrar un mensaje personalizado.
✉️ Libro 'The Great Gatsby' devuelto con éxito.

✉️ Este libro no está prestado.
📋 Mostrar los libros disponibles:

Mostrar un mensaje indicando los libros disponibles para préstamo.
📚 Libros disponibles para préstamo:
- 'Ready Player One'
- 'Top Gun: Maverick'
- 'Gran Turismo'
### ✅ Desafío adicional: 
Considera manejar situaciones como intentos de préstamo de libros inexistentes o la actualización de detalles de libros existentes.

```python
biblioteca = {
    'a1b2': {'titulo': 'Feminismo para principiantes', 'autor': 'Nuria Varela', 'año': 2006},
    'a1b3': {'titulo': 'Kafka en la orilla', 'autor': 'Karuki Murakami', 'año': 2003},
    'a1b4': {'titulo': 'Demian', 'autor': 'Herman Hesse', 'año': 1919},
    'a1b5': {'titulo': 'La vegetariana', 'autor': 'Hang Kang', 'año': 2022},
    'a1b6': {'titulo': 'Los recuerdos del Porvenir', 'autor': 'Elena Garro', 'año': 1963},
    'a1b8': {'titulo': 'La policia de la memoria', 'autor': 'Yoko Ogawa', 'año': 1994}
}

prestados = set()

while True: 
    print(
        """
        Biblioteca
        1. Prestar libro
        2. Devolver libro
        3. Ver libros disponibles
        4. Salir
        5. Actualizar libro
        """
    )

    opcion = int(input("Selecciona una opción: "))

    match opcion:
        case 1: 
            codigo = input("Código del libro: ").lower()

            if codigo not in biblioteca:
                print("El libro no existe.")
            
            elif codigo in prestados:
                print("El libro que seleccionaste ya ha sido prestado.")
            else:
                prestados.add(codigo)
                print(f"Libro: '{biblioteca[codigo]['titulo']}' prestado con éxito.")

        case 2: 
            codigo = input("Código del libro: ").lower()

            if codigo in prestados:
                prestados.remove(codigo)
                print(f"Libro '{biblioteca[codigo]['titulo']}' devuelto con éxito.")
            else: 
                print("Este libro no ha sido prestrado.")

        case 3: 
            print("\nLibros disponibles: \n")

            for codigo, libro in biblioteca.items():
                
                if codigo not in prestados:
                    print(f"{codigo} - {libro['titulo']} ({libro['autor']})")
        
        case 5:
            codigo = input("Código del libro a actualizar: ").lower()

            if codigo in biblioteca:
                biblioteca[codigo]["titulo"] = input("Nuevo título: ")
                biblioteca[codigo]["autor"] = input("Nuevo autor: ")
                biblioteca[codigo]["año"] = input("Nuevo año: ")

                print("Libro actualizado.")
            else: 
                print("El libro no existe.")

        case 4:
            print("Gracias por utilizar nuestra biblioteca.")
            break

        case _:
            print("Opción inválida.")
```
