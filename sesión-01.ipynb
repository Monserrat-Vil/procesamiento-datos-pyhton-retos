## Reto 01: Sistema de Recomendación de Cursos

### 🎯 Objetivo

Disponibilidad para inscripción (Boolean).

```python
#Principiante
nombre_curso_1 = "Introducción a Python"
codigo_curso_1 = 101
duracion_curso_1 = 45.5
disponibilidad_curso_1 = True
print(type(nombre_curso_1))
print(type(codigo_curso_1))
print(type(duracion_curso_1))
print(type(disponibilidad_curso_1))

nombre_curso_2 = "Álgebra lineal"
codigo_curso_2 = 109
duracion_curso_2 = 60.2
disponibilidad_curso_2 = False

nombre_curso_3 = "Introducción a Ciencia de Datos"
codigo_curso_3 = 205
duracion_curso_3 = 35.6
disponibilidad_curso_3 = True

print(f"Curso: {nombre_curso_1} - Código: {codigo_curso_1} - Duración: {duracion_curso_1} horas - Disponible: {disponibilidad_curso_1}")
print(f"Curso: {nombre_curso_2} - Código: {codigo_curso_2} - Duración: {duracion_curso_2} horas - Disponible: {disponibilidad_curso_2}")
print(f"Curso: {nombre_curso_3} - Código: {codigo_curso_3} - Duración: {duracion_curso_3} horas - Disponible: {disponibilidad_curso_3}")

print("-"*80)
#Intermedio
cursos = [
    {"nombre": "Introducción a Python", "codigo": 101, "duracion": 45.5, "disponible": True,},    
    {"nombre": "Álgebra lineal", "codigo": 109, "duracion": 60.2, "disponible": False,},
    {"nombre": "Introducción a Ciencia de Datos", "codigo": 205, "duracion": 35.6, "disponible": True},
    {"nombre": "SQL para Análisis de Datos", "codigo": 301, "duracion": 40.0, "disponible": True},
    {"nombre": "Machine Learning", "codigo": 410, "duracion": 75.0, "disponible": False}
    ]

for curso in cursos:
    print(type(curso["nombre"]))
    print(type(curso["codigo"]))
    print(type(curso["duracion"]))
    print(type(curso["disponible"]))

    print(
        f"Curso: {curso["nombre"]} - "
        f"Código: {curso["codigo"]} - "
        f"Duración: {curso["duracion"]} - "
        f"Disponibilidad: {curso["disponible"]}"
    )

```

## Reto 02: Simulador de compra de artículos

### 🎯 Objetivo

### 📝 Instrucciones
🛒 Identifica 5 productos distintos con los siguientes datos:

Nombre del producto.
Precio.
Stock.
Cantidad.
⌨️ Declara las variables necesarias para almacenar la información de los productos, incluyendo su nombre, precio, stock y cantidad de cada producto que deseas comprar.

Puedes usar asignación múltiple para declarar las variables de los productos, en una sola línea:

nombre, precio, stock, cantidad = "Producto 1", 10.0, 5, 2
🧮 Realiza las siguientes comprobaciones usando operadores aritméticos, lógicos y relacionales:

Comparar si el stock de un producto es mayor que el de otro.
Verificar si un producto específico tiene un precio mayor a 20 y su stock es menor a 5.
Comprobar si dos productos tienen el mismo nombre.
Evaluar si el precio de un producto multiplicado por su cantidad es mayor que 100.
Comparar si la cantidad de un producto es mayor a 10.
Aplicar un descuento del 15% en el precio de un producto utilizando operadores de asignación.
🖨️ Imprime los resultados de cada una de las condiciones evaluadas.

💰 Calcula y almacena el total a pagar por los cinco productos.

📊 Finalmente, imprime el total a pagar de la siguiente manera:

🛍️ Total a pagar por los productos es: $XX.XX

```python
#Principiante
producto_1 = "Celular", 7000.00, 12, 2
producto_2 = "Laptop", 10000.00, 5, 1
producto_3 = "Televisión", 5000.00, 15, 3
producto_4 = "Audífonos", 500.00, 60, 1
producto_5 = "Smartwatch", 1000.00, 39, 7

print(
    f"1. Comparar si el stock de un producto es mayor que el de otro: {producto_1[2] > producto_2[2]}\n" 
    f"2. Verificar si un producto específico tiene un precio mayor a 20 y su stock es menor a 5: {(producto_3[1] > 20) and (producto_3[2] < 5)}\n"
    f"3. Comprobar si dos productos tienen el mismo nombre. {producto_4[0] == producto_5[0]}\n"
    f"4. Evaluar si el precio de un producto multiplicado por su cantidad es mayor que 100. {producto_1[1]*producto_1[3] > 100}\n"
    f"5. Comparar si la cantidad de un producto es mayor a 10. {producto_4[3] > 10}\n"
    f"6. Aplicar un descuento del 15% en el precio de un producto utilizando operadores de asignación. {producto_3[1] * 0.85}"
    )

cuenta = (producto_1[1] * producto_1[3]) + (producto_2[1] * producto_2[3]) + (producto_3[1]*0.85 * producto_3[3]) + (producto_4[1] * producto_4[3]) + (producto_5[1] * producto_5[3]) 

print(f"Total a pagar: ${cuenta:.2f}")

print("-"*80)

#Intermedio

#Lista de productos

productos = [
    {"nombre": "Celular", "precio": 7000.00, "stock": 12, "cantidad": 2},
    {"nombre": "Laptop", "precio": 10000.00, "stock": 5, "cantidad": 1},
    {"nombre": "Televisión", "precio": 5000.00, "stock": 15, "cantidad": 3},
    {"nombre": "Audífonos", "precio": 500.00, "stock": 60, "cantidad": 1},
    {"nombre": "Smartwatch", "precio": 1000.00, "stock": 39, "cantidad": 7}
]

# Operadores relacionales y lógicos
print(f"1. ¿El stock del Celular es mayor que el de la Laptop? "
    f"{productos[0]['stock'] > productos[1]['stock']}")

print(f"2. ¿La Televisión cuesta más de $20 y tiene menos de 5 unidades? "
    f"{productos[2]['precio'] > 20 and productos[2]['cantidad'] > 5}"
)

print(f"3. ¿Audífonos y Smartwatch tienen el mismo nombre? "
    f"{productos[3]['nombre'] == productos[4]['nombre']}"
    )

print(f"4. ¿El precio de celular multiplicado por su cantidad es mayor que 100? " 
    f"{productos[0]['precio']*productos[0]['cantidad'] > 100}"
    )
    
print(f"5. ¿La cantidad de Smartwatchs es mayor a 10?? "
    f"{productos[3]['cantidad'] > 10}"
    )

# Operador de asignación
productos[2]['precio'] *= 0.85
    
print(f"6. Precio de la televisión con 15 % de descuento: "
    f"${productos[2]['precio']:.2f}"
)


#Total de la compra

total = sum(
    producto["precio"] * producto["cantidad"]
    for producto in productos
)

print(f"\n Total a pagar por los productos es: ${total:.2f}")
```

## Reto 03: Cotización y financiamiento de un automóvil

### 🎯 Objetivo
⚒️ Desarrollar un programa en Python que simule la cotización y financiamiento de un automóvil, evaluando diferentes condiciones para calcular la cuota mensual y mostrar una tabla de amortización simple. Este ejercicio te permitirá practicar el uso de operadores de asignación, interpolación de strings y lectura por teclado en Python.

### 📝 Instrucciones

🧮 Realiza las siguientes comprobaciones utilizando operadores de asignación y aritméticos:

🖨️ Muestra los detalles del financiamiento utilizando interpolación de strings:

📊 Genera y muestra una tabla de amortización simple para los primeros dos meses:

✨ Imprime un mensaje de despedida agradeciendo al usuario por utilizar el sistema de financiamiento.

🚀 Este es un ejemplo de como seria el resultado.

🚗 Modelo: Toyota Corolla
💰 Precio total: $50,000.00
📅 Meses de financiamiento: 12
💳 Cuota mensual: $4,166.67

📊 Tabla de amortización
-------------------------
Mes 1: Pago $1,666.67, Saldo pendiente $18,333.33
Mes 2: Pago $1,666.67, Saldo pendiente $16,666.67

¡Gracias por utilizar nuestro sistema de financiamiento!
✅ Desafío adicional: intenta imprimir una tabla de amortización completa para todos los meses del financiamiento, mostrando el pago mensual y el saldo pendiente para cada mes.

```python
#Principiante 
modelo = input("🚗 Introduce el modelo del automóvil que deseas comprar: ")
precio = float(input("Coloca el precio estimado del auto: $ "))
meses = int(input("Selecciona el número de meses para el financiamiento. (12, 24, 36, 48, 60): "))
cuota_mensual = precio / meses

print(
    f"\nModelo: {modelo} "
    f"\nPrecio total: {precio} "
    f"\nMeses de financiamiento: {meses} "
    f"\nCuota mensual: {cuota_mensual} "
    )

print("\n📊 Tabla de amortización: ")
 
saldo = precio

for mes in range( 1, meses + 1):
    saldo -= cuota_mensual

    if saldo < 0:
        saldo = 0

    print(
        f"Mes {mes:>2}: "
        f"Pago ${cuota_mensual:>10,.2f} | "
        f"Saldo pendiente ${saldo:>10,.2f}"
    )

print("\n¡Gracias por utilizar nuestro sistema de financiamiento!")
```
